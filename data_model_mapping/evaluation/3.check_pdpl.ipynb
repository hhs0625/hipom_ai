```python
import pandas as pd
import matplotlib.pyplot as plt
import json

file_name = 'train_t5-base_td_unit_80/checkpoint-4880/test_with_predictions'
# Load the data
df_test = pd.read_csv(f"{file_name}.csv")


# Load the ships_idx_selection.json file
with open('../translation/ship_idx_selection.json', 'r') as json_file:
    ship_idx_selection = json.load(json_file)

train_idx = set(ship_idx_selection["training_ship_idx_selection"])
val_idx = set(ship_idx_selection["validation_ship_idx_selection"])
test_idx = set(ship_idx_selection["test_ship_idx_selection"])

mdm_true_df = df_test[df_test['MDM']]
grouped = mdm_true_df.groupby('ships_idx')['sdl_tp_correct'].mean().reset_index()
grouped['sdl_tp_correct'] = grouped['sdl_tp_correct'] * 100  # Convert to percentage

# Calculate the overall average
overall_average = grouped['sdl_tp_correct'].mean()

# Print the overall average
print(f"Overall Average Percentage of Correct sdl_tp_correct: {overall_average:.2f}%")

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(grouped['ships_idx'], grouped['sdl_tp_correct'], color='skyblue')
plt.xlabel('Ships Index')
plt.ylabel('Percentage of Correct sdl_tp_correct')
plt.title('Percentage of Correct sdl_tp_correct per Ships Index for MDM=True')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
