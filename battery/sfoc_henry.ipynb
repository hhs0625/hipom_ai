```python
import pandas as pd
import matplotlib.pyplot as plt
newest_csv_file_path = 'MNA_Curve.csv'
newest_data = pd.read_csv(newest_csv_file_path)

plt.figure(figsize=(10, 6))
plt.plot(newest_data['Engine Load'], newest_data['Relative Fuel Consumption Rate'], color='red', label='MAN')
plt.xlabel('Engine Load [Relative units]')
plt.ylabel('Relative Fuel Consumption Rate [Relative units]')
plt.title('Relative Fuel Consumption Rate vs Engine Load for MAN engines')
plt.ylim(0.9, 1.4)
plt.legend()
plt.grid(True)
plt.show()
```
